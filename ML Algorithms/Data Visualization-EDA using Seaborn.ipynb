# Seaborn

## plots for continuous variables/ features

```python
import seaborn as sns
data=sns.load_dataset("tips")
```

```python
data.head()
```

```python
data.info()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="scatter")
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="reg")
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,hue="sex")
```

```python
data['smoker'].value_counts()
```

```python
sns.pairplot(data,hue="smoker")
```

```python
sns.distplot(data["tip"],kde=True)
```

```python
sns.distplot(data["tip"],kde=False)
```

## plots for categorical features

```python
sns.countplot(data['sex'],data=data)
```

```python
sns.countplot(data['smoker'],data=data)
```

```python
sns.countplot(y='sex',data=data)
```

```python
sns.countplot(data['day'],data=data)
```

```python
sns.barplot(x='total_bill',y='sex',data=data)
```

```python
sns.boxplot(x='sex',y='total_bill',data=data)
```

```python
sns.boxplot(x='day',y='total_bill',data=data,palette='rainbow')
```

```python
sns.boxplot(data=data,orient='v')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=data)
```

```python
sns.violinplot(x='total_bill',y='day',data=data,palette='rainbow')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,2)
plt.plot(x,y,'yellow')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# remove missing values from dataframe
cars.dropna(axis=0,inplace=True)
```

```python
cars.shape
```

```python
# Scatterplot
plt.scatter(cars['Age'],cars['Price'],c='blue')
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Scatterplot of Price vs Age')
plt.show()
plt.savefig('cars prive for Age.png')
```

```python
plt.bar(cars['HP'],cars['Price'])
plt.xlabel('FuelType')
plt.ylabel('Price')
plt.title('Scatterplot of Price vs FuelType')
plt.show()
```

```python
# Histogram: to plot frequency of numerical attribute
plt.hist(cars['Age'],bins=20)
plt.title('Histogram of Kilometer')
plt.xlabel('kilometer')
plt.ylabel('frequency')
plt.show()
```

```python
# scatterplot using seaborn
import seaborn as sns
sns.scatterplot(data=cars,x=cars['Age'],y=cars['Price'],hue=cars['FuelType'],size=cars['FuelType'])
```

```python
sns.distplot(cars['Age'],bins=10,norm_hist=False,kde=True)
```

```python
sns.countplot(y=cars['FuelType'],data=cars)
```

```python
# Grouped Bar Plot
sns.countplot(x=cars['FuelType'],data=cars,hue=cars['Automatic'])
```
